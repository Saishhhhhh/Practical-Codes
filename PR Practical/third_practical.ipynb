```python
import matplotlib.pyplot as plt

def generate_pattern(input_string):
    pattern = ""
    prev_char = ""
    count = 1
    freq_dict = {}

    for char in input_string:
        if char == prev_char:
            count += 1
        else:
            if prev_char:
                pattern += prev_char + str(count)
                freq_dict[prev_char] = count
            prev_char = char
            count = 1

    if prev_char:
        pattern += prev_char + str(count)
        freq_dict[prev_char] = count

    return pattern, freq_dict

def visualize_pattern(freq_dict):
    characters = list(freq_dict.keys())
    counts = list(freq_dict.values())

    plt.figure(figsize=(8, 5))
    bars = plt.bar(characters, counts, color='skyblue', edgecolor='black')
    plt.title("Character Frequency Pattern - Visual")
    plt.xlabel("Characters")
    plt.ylabel("Count")

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.1, int(yval), ha='center', va='bottom')

    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("\n\n")
    print("Name:- Sudhir Kumar Singh \n")
    print("Roll no:- TEAD23152 \n")
    input_string = input("Enter a string: ")
    pattern, freq_dict = generate_pattern(input_string)
    print("Generated Pattern:", pattern)

    # Show visualization
    visualize_pattern(freq_dict)

    # aaabbc

```
